## Creating a function to retrieve information from the API (see Spotipy documentation) 

```python
# testing to see which features are available so that I can assign later on to df 
uri = 'spotify:track:0VjIjW4GlUZAMYd2vXMi3b'
results_1 = sp.track(uri)
results_2 = sp.audio_features(uri)

# view the different features we want for results_1
# results_1["name"]
# results_1['album']['name']
# results_1['album']['release_date']
# results_1['duration_ms']
# results_1['popularity']

# viewing the different features we want for results_2, we use 0 index position because it is a dict in a list 
# results_2[0]['tempo']
# results_2[0]['danceability']
# results_2[0]['loudness']
```

```python
# creating a function that we can reuse when wanting to get the info from the given track in Hot 100 Songs 
# we pass the URI from the song as an argument. Then we assign the variable for the feature we receive from using sp.track and sp.track_audio
# lastly we store that in a list called track and return it for when we are requesting the songs we want 

def getTrackInfo(uri):
  track_info = sp.track(uri)
  track_audio = sp.audio_features(uri)

  # these variables come from using the .track request
  song = track_info['name']
  album_name = track_info['album']['name']
  date_of_release = track_info['album']['release_date']
  # will convert this to minutes later on 
  duration = track_info['duration_ms'] 
  popularity = track_info['popularity']

  tempo = track_audio[0]['tempo']
  danceability = track_audio[0]['danceability']
  loudness = track_audio[0]['loudness']

  track =  [song, album_name, date_of_release, duration, popularity, tempo, danceability, loudness]
  return track
```
